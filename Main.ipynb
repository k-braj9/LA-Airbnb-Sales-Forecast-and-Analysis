# Modeling Monthly Revenue

```python
# How much Airbnb revenue will be generated next month?
```

```python
monthly_revenue = test_data.groupby('month')['price'].sum().reset_index()
```

```python
monthly_revenue
```

```python
train_split = int(len(monthly_revenue) * 0.8)

train_old = monthly_revenue.iloc[:train_split]
test_old = monthly_revenue.iloc[train_split:]

from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_old['price'], order=(1,1,0))
model_old = model.fit()

forecast_old = model_old.forecast(steps=len(test_old))

from sklearn.metrics import mean_absolute_error
mae_old = mean_absolute_error(test_old['price'], forecast_old)
mae_old
```

```python
#The ARIMA model achieved a mean absolute error (MAE) of approximately 31,600 against an average monthly revenue of 31,620, corresponding to an average forecasting error of about 5.7%. This indicates strong predictive performance for short-term revenue forecasting.
```

```python
forecast_new = model_new.forecast(steps=len(test_new))

from sklearn.metrics import mean_absolute_error
mae_new = mean_absolute_error(test_new['price'], forecast_new)
mae_new
```

```python
plt.plot(test_old['month'], forecast_old, label='Before Improvement')
plt.plot(test_new['month'], forecast_new, label='After Improvement')
plt.legend()
plt.show()
```
